```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(
    './sample_superstore.xls',
    sheet_name='Orders'
)
```

```python
df[:1000]
```

```python
sns.relplot(
    x='Sales',
    y='Profit',
    hue='Profit',
    height=10,
    palette="RdYlGn",
    data=df,
)

sns.lineplot(
    x=[0,25_000],
    y=[0, 2500],
    legend=None,
    color="green",
    linewidth=2
)

plt.show()
```
